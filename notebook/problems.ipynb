## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Normal Distribution
mean = 0
sd = 1
x = np.linspace(mean - 4*sd, mean + 4*sd, 1000)

# Histogram with Overlay
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm  # Ensure norm is imported from scipy.stats

# Parameters for Normal Distribution
mean = 0
sd = 1

# Generate random samples
samples = norm.rvs(loc=mean, scale=sd, size=1000)

# Plot histogram of samples
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='orange', edgecolor='black', label="Histogram (Samples)")

# Overlay PDF on histogram
x_pdf = np.linspace(min(samples), max(samples), 100)
pdf_samples = norm.pdf(x_pdf, loc=mean, scale=sd)
plt.plot(x_pdf, pdf_samples, label="PDF", color="red", linewidth=2)

# Add labels and title
plt.title("Random Samples from Normal Distribution with PDF Overlay")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()

# PDF Plot
pdf = norm.pdf(x, loc=mean, scale=sd)
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label="PDF", color="blue")
plt.fill_between(x, pdf, alpha=0.2, color="blue")
plt.title("Normal Distribution - PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()

# CDF Plot
cdf = norm.cdf(x, loc=mean, scale=sd)
plt.figure(figsize=(10, 6))
plt.plot(x, cdf, label="CDF", color="green")
plt.title("Normal Distribution - CDF")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.grid()
plt.show()

# Quantile Function Plot
quantiles = np.linspace(0.01, 0.99, 100)
q_values = norm.ppf(quantiles, loc=mean, scale=sd)
plt.figure(figsize=(10, 6))
plt.plot(quantiles, q_values, label="Quantile Function", color="purple")
plt.title("Normal Distribution - Quantile Function")
plt.xlabel("Probability (Quantiles)")
plt.ylabel("Quantile Value")
plt.legend()
plt.grid()
plt.show()

# Random Variable Generation Plot
def rnorm(n, mean=0, sd=1):
    """
    Generate n random values from a normal distribution with specified mean and standard deviation.
    
    Parameters:
        n (int): Number of random values to generate.
        mean (float): Mean of the normal distribution (default is 0).
        sd (float): Standard deviation of the normal distribution (default is 1).
        
    Returns:
        np.ndarray: Array of random values from the normal distribution.
    """
    return np.random.normal(loc=mean, scale=sd, size=n)

# Example usage of rnorm function
n_samples = 1000
mean = 0
sd = 1

# Generate random samples using rnorm
samples = rnorm(n_samples, mean=mean, sd=sd)

# Plot histogram of generated samples
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='orange', edgecolor='black', label="Histogram (Samples)")

# Overlay PDF on histogram for comparison
x_pdf = np.linspace(min(samples), max(samples), 100)
pdf_samples = norm.pdf(x_pdf, loc=mean, scale=sd)
plt.plot(x_pdf, pdf_samples, label="PDF", color="red", linewidth=2)

# Add labels and title
plt.title("Normal Distribution with Random Variable Generation")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```
